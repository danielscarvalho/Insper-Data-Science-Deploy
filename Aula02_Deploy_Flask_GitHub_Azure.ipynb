**Criando um Serviço de aplicativo no Azure**

Neste estudo vamos utilizar o App Service ou Serviços de Aplicativos, que é categorizado como um serviço como plataforma (PaaS - Plataform as a Service). Neste tipo de serviço o desenvolvedor deve se preocupar  com a aplicação e seus dados, deixando a responsabilidade de gerenciar a rede, sistema operacional, virtualização e outros pontos para a Azure.


1.	Após fazer login, clique em "Criar um recurso" no canto superior esquerdo e procure **Serviço de aplicativos**.



2.	Selecione **Aplicativo Web** na lista de resultados e clique em **Criar** na página seguinte.






3. Criando o  **SERVIÇO - APLICATIVO WEB**

* Na página de criação do aplicativo, preencha as informações básicas, como assinatura, grupo de recursos, nome do App e região.
* Em "Pilha de Runtime", selecione "Python".
* Escolha a versão desejada do Python.





* **ESCOLHA A OPÇÃO GRATUITA DE PLANO**

<h5 style="color:brown"> ATENÇÃO AO ESCOLHER O PLANO - ESCOLHA A OPÇÃO GRATUITA DE PLANO </h5> 




* **CLique em Revisar + Criar**


4. Com o aplicativo CRIADO. Dê um duplo clique para acessar os detalhes de configuração do aplicativo web.




**Configuração do Centro de Implantação**


* Na opção de **Source** selecione **GitHub** como fonte de código

* Clique em "Autorizar" e siga as instruções para conectar sua conta do GitHub ao Azure.
* Após a autorização, você poderá selecionar o repositório do GitHub contendo a aplicação Flask que deseja implantar.
* Escolha o branch apropriado e o caminho para a pasta raiz do projeto Flask.


* Finalizada a configuração, clique no botão **Save**. Automaticamente será criado um BUILD e DEPLOY na área de ACTIONS do GitHub.


**ATENÇÃO**  
O Azure executa o arquivo que esteja como app.py. No caso do exemplo, o arquivo a ser executado é o core e app. Então, é necessário entrar na 
seção **Configuração**. 

**Concluída a configuração podemos executar o teste**
