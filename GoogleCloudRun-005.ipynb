### O que é um contêiner (Docker)?

Docker é uma plataforma que permite criar, implantar e executar aplicações em contêineres, que são ambientes leves e isolados, garantindo consistência entre desenvolvimento, testes e produção.

O Docker é uma plataforma aberta para desenvolver, enviar e executar aplicações (em Python Flask por exemplo). O Docker permite separar as aplicações da infraestrutura, possibilitando entregar software rapidamente. Com o Docker, podemos gerenciar infraestrutura da mesma forma que gerenciamos as aplicações, com arquivos de configuração. Ao aproveitar as metodologias do Docker para enviar, testar e implantar (**deploy**) código, podemos reduzir significativamente o atraso entre escrever o código e executá-lo na produção. \[Docker\]

Simplificando, um contêiner é um processo em área restrita (sandboxed) em sua máquina (local ou em nuvem) que é isolado de todos os outros processos na máquina host. Esse isolamento aproveita namespaces e cgroups do kernel, recursos que estão no Linux há muito tempo. O Docker trabalhou para tornar esses recursos acessíveis e fáceis de usar. Para resumir, um contêiner:

- É uma instância executável de uma imagem. 
- Você pode criar, iniciar, parar, mover ou excluir um contêiner usando o DockerAPI ou CLI.
- Pode ser executado em máquinas locais, máquinas virtuais ou implantado na nuvem.
- É portátil (pode ser executado em qualquer sistema operacional)
- Os contêineres são isolados uns dos outros e executam seus próprios softwares, binários e configurações.
- A partir de uma imagem, díversas instâncias podem ser executadas em um mesmo ambiente, ou diversos ambientes distintos (distribuídos)

### O que é uma imagem de contêiner?

Ao executar um contêiner, ele usa um sistema de arquivos isolado. Esse sistema de arquivos personalizado é fornecido por uma imagem de contêiner. Como a imagem contém o sistema de arquivos do contêiner, ela deve conter tudo o que é necessário para executar uma aplicação - todas as dependências, configurações, scripts, binários, etc. A imagem também contém outras configurações para o contêiner, como variáveis de ambiente, um comando padrão a ser executado, e outros metadados.

**Docker vs VM (Virtual Machine)**

Uma máquina virtual (VM) e um Docker são tecnologias de virtualização, mas operam em níveis diferentes: uma VM emula um sistema operacional completo, incluindo kernel, bibliotecas e aplicativos, sobre um hipervisor (VMWare, Oracle Virtual Box, Hyper-V, etc...), o que a torna mais pesada e lenta, já que cada VM carrega uma instância do SO inteiro; já o Docker utiliza contêineres que compartilham o kernel do sistema operacional hospedeiro (host), contendo apenas as bibliotecas e dependências necessárias para a aplicação, tornando-o mais leve, rápido e eficiente em recursos, ideal para implantações consistentes e escaláveis, enquanto VMs são mais indicadas para isolamento total ou execução de sistemas operacionais distintos.

## Serviços Serverless de Docker em Nuvem: GCP, AWS e Azure

Abaixo, explico os principais serviços serverless que utilizam Docker (ou contêineres) nas três grandes plataformas de nuvem — Google Cloud Platform (GCP), Amazon Web Services (AWS) e Microsoft Azure — e apresento uma tabela comparativa. Esses serviços permitem executar contêineres sem gerenciar a infraestrutura subjacente, ideal para aplicações modernas baseadas em microserviços.

## Serviços Serverless de Docker nas Plataformas

### 1. Google Cloud Platform (GCP): Cloud Run
- **Descrição**: O Cloud Run é a principal oferta serverless da GCP para contêineres Docker. Ele permite que você implante contêineres HTTP stateless (sem estado) que são automaticamente escalados com base na demanda. Você carrega uma imagem Docker, e o Cloud Run gerencia a execução, escalonamento e infraestrutura.
- **Características**:
  - Suporte nativo a contêineres Docker.
  - Escalonamento automático (de zero a milhares de instâncias).
  - Integração com o Google Container Registry (GCR) para armazenar imagens.
  - Suporte a múltiplas linguagens, já que o contêiner pode rodar qualquer runtime.
  - Focado em simplicidade e experiência do desenvolvedor, com deploy rápido via comandos simples.
  - Limitação: Suporte oficial apenas para JavaScript no Cloud Functions (outro serviço serverless da GCP), mas no Cloud Run, qualquer linguagem é suportada via contêineres.
- **Casos de Uso**: APIs REST, aplicações web, microsserviços, e workloads event-driven.

### 2. Amazon Web Services (AWS): AWS Fargate
- **Descrição**: O AWS Fargate é um serviço serverless para contêineres Docker, integrado ao Amazon Elastic Container Service (ECS) e ao Elastic Kubernetes Service (EKS). Ele elimina a necessidade de gerenciar servidores ou clusters, permitindo que você foque na execução de contêineres.
- **Características**:
  - Suporte a contêineres Docker via ECS ou EKS.
  - Escalonamento automático com base em políticas definidas.
  - Integração com o Amazon Elastic Container Registry (ECR) para imagens.
  - Controle granular de recursos (CPU e memória) para cada contêiner.
  - Suporte a redes seguras via Amazon VPC.
  - Mais complexo de configurar em comparação com o Cloud Run, mas oferece maior controle.
- **Casos de Uso**: Aplicações em contêineres que requerem integração profunda com o ecossistema AWS, como pipelines CI/CD ou workloads Kubernetes.

### 3. Microsoft Azure: Azure Container Instances (ACI)
- **Descrição**: O Azure Container Instances (ACI) é o serviço serverless da Azure para executar contêineres Docker sem gerenciar a infraestrutura. Ele é simples e permite rodar contêineres isolados ou em grupos (container groups) com um único comando.
- **Características**:
  - Suporte a contêineres Docker com deploy rápido.
  - Escalonamento manual (não automático como Cloud Run ou Fargate, mas pode ser integrado com outros serviços para automação).
  - Integração com o Azure Container Registry (ACR).
  - Suporte a execução de comandos em contêineres via shell interativo para debugging.
  - Pode rodar em Windows ou Linux.
  - Menos foco em escalonamento automático, mais voltado para cenários de execução sob demanda ou testes.
- **Casos de Uso**: Testes rápidos, tarefas temporárias, ou aplicações simples que não requerem orquestração complexa.

## Tabela Comparativa

| **Aspecto**               | **GCP: Cloud Run**                          | **AWS: Fargate**                          | **Azure: Container Instances (ACI)**      |
|---------------------------|---------------------------------------------|-------------------------------------------|-------------------------------------------|
| **Modelo Serverless**     | Totalmente serverless, com foco em HTTP     | Serverless, integrado a ECS/EKS           | Serverless, mas com menos automação       |
| **Suporte a Docker**      | Sim, nativo                                 | Sim, via ECS ou EKS                       | Sim, nativo                               |
| **Escalonamento**         | Automático (de 0 a milhares)                | Automático, com políticas definidas       | Manual (pode ser automatizado com outros serviços) |
| **Integração com Registry**| Google Container Registry (GCR)            | Amazon Elastic Container Registry (ECR)   | Azure Container Registry (ACR)            |
| **Linguagens Suportadas** | Qualquer (via contêiner)                    | Qualquer (via contêiner)                  | Qualquer (via contêiner)                  |
| **Facilidade de Uso**     | Muito alta (deploy com um comando)          | Moderada (mais configuração necessária)   | Alta (simples, mas menos automação)       |
| **Controle de Recursos**  | Básico (CPU/memória configuráveis)          | Avançado (controle granular)              | Básico (CPU/memória configuráveis)        |
| **Casos de Uso**          | APIs, microsserviços, workloads HTTP        | Workloads complexas, Kubernetes           | Testes, tarefas temporárias, execuções simples |
| **Rede e Segurança**      | Integração com VPC, HTTPS/TLS               | Integração com Amazon VPC, segurança robusta | HTTPS/TLS, integração com redes virtuais  |

## Análise Comparativa
- **Facilidade de Uso**: O Cloud Run (GCP) é o mais simples para iniciantes, com deploy rápido e foco em workloads HTTP. O ACI (Azure) também é fácil, mas menos automatizado. O Fargate (AWS) exige mais configuração, especialmente se usado com EKS.
- **Escalonamento**: Cloud Run e Fargate oferecem escalonamento automático robusto, enquanto o ACI é mais limitado nesse aspecto, sendo mais adequado para tarefas pontuais.
- **Flexibilidade**: Fargate oferece mais controle e integração com o ecossistema AWS, ideal para cenários complexos. Cloud Run e ACI são mais simples, mas menos flexíveis.
- **Ecossistema**: AWS Fargate se beneficia do vasto ecossistema AWS, enquanto Cloud Run e ACI têm integração sólida com seus respectivos ecossistemas (GCP e Azure).

## Conclusão
- **Escolha o Cloud Run (GCP)** se você busca simplicidade, deploy rápido e workloads HTTP escaláveis.
- **Escolha o Fargate (AWS)** se precisa de controle granular, integração com Kubernetes, ou já está no ecossistema AWS.
- **Escolha o ACI (Azure)** para tarefas temporárias, testes, ou se você já usa outros serviços Azure e não precisa de escalonamento automático avançado.

Adicionar as chaves de criptografia do site oficial do Docker (GPG key):

```bash
 sudo mkdir -p /etc/apt/keyrings

 curl -fsSL https://download.docker.com/linux/ubuntu/gpg | sudo gpg --dearmor -o /etc/apt/keyrings/docker.gpg
```

Adicionar (append) o endereço do repositório do Docker na lista de respositórios do Linux:

```bash
     echo \
      "deb [arch=$(dpkg --print-architecture) signed-by=/etc/apt/keyrings/docker.gpg] https://download.docker.com/linux/ubuntu \
      $(lsb_release -cs) stable" | sudo tee /etc/apt/sources.list.d/docker.list > /dev/null
```

Atualizar a lista de programas disponíveis nos repositórios e instalar o Docker Engine:


1. Criar uma pasta do projeto com o arquivo texto Dockerfile e todos os arquivos (scripts) que compõe o projeto

2. Subir a imagem criando uma instância do Docker (contâiner)

### CI/CD - Continuous Integration/Continuous Delivery

Com o **Google Cloud Run** podemos atualizar de forma automatizada o contêiner, conforme o código da aplicação é atualizado no GitHub!

Desta forma, quando o código do GitHub é atualizado na branch master (main), o Google Cloud Run faz um novo build da imagem e carrega em produção (deploy).

Principais serviços do Google Cloud, há equivalencia de serviços em outras plataformas

VM:

- **Google Cloud Engine**: Fornecimento de máquina virtual (VM), vimos equivalente no Microsoft Azure e no Oracle Cloud

Serverless:

- **Google Cloud App**: Subimos os scripts Python fom Flask para fornecer WEB APIs (Serviços WEB)
- **Google Cloud Run**: Subimos os scripts Python com Flask em um container
- **Google Cloud Functions**: Trata-se de microserviços, onde as funções de um script são fornecidos como WEB API (não vimos neste workshop)

### Docker no ambiente Microsoft Azure

Podemos subir nossa imagem Docker local no repositório do Docker Hub:

Para se conectar ao repo Docker Hub online:

```bash
docker login
```

Para fazer o build (gerar imagem) usando o nome do seu repo no Docker Hub:

```bash
docker build -t danielscarvalho/insper-ds-lab-3 .
```

Para enviar a imagem para o repo no Docker Hub:

```bash
docker push danielscarvalho/insper-ds-lab-3
```

No Azure Contênier Instance, você cria um novo Docker com base na imagem que foi carregada (push) no Docker Hub, usar o nome do repositório:
